# Review: Interactions in a single species

Recall the example of trout in a lake. The lake has a carrying vcapacity of 4000 and the trout grow with a rate of 15%. We can model the rate of change of the trout population ($P_1$) as:

$$\dot{P} = r_1P_1 + -r_1 \left(\frac{P_1}{M}\right)P_1$$

where $r_1$ is the rate of growth (15%), $M$ is the carying capacity, and $P_1$ is the trout population. You can think about the carying capacity term as an interaction of the population with the enviroment itself.

# Interactions between populations

Now we would like to add a second fish to the lake. We will call this fish two and depote its population as $P_2$. The two populations will have an interaction with each other since they are occupying the same space and are competing for the same resources.

We model this interaction with an **interaction term** that depends both on the population of the first fish ($P_1$) and the population of the second fish ($P_2$). In the case of a single fish, this is analagous to the carrying capacity term. In our current competition model the population for fish 1 is given by:

$$\dot{P_1} = \text{rate of change from births} + \text{rate of change from deaths}$$
$$\dot{P_1} = +r_{b,1}P_1 + -r_{d,1}P_1P_2$$

Where $r_{b,1}$ is the rate of births for the first fish population, and $r_{d,1}$ is the rate of death for the first fish population.
